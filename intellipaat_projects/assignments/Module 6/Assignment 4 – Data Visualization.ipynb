1. Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data
frame.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv(r"csv files\cars-4.csv")
df.head()
```

```python
df['cyl'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10,10))
df['model'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

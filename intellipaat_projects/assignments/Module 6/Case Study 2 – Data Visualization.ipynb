```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df = pd.read_csv(r"csv files\cars-6.csv")
df.head()
```

```python
sns.barplot(y=df['cyl'].value_counts(),x=df['cyl'].value_counts().keys())
plt.xlabel('Number of cylinders')
plt.ylabel('Frequency of cars')
plt.show()
```

```python
df.isna().sum()
```

```python
sns.barplot(x=df.isna().sum().keys(), y=df.isna().sum())
plt.show()
```

```python
sns.histplot(df['mpg'])
plt.xlabel('Miles per gallon')
plt.ylabel('Frequency')
plt.title('Miles Per Gallon Histogram')
plt.show()
```

```python
sns.boxplot(df['hp'])
plt.xlabel('Car Horsepower')
plt.title('Boxplot for car horsepower')
plt.show()
```
